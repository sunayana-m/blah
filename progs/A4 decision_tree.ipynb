```python
import pandas as pd
import numpy as np
data=pd.read_csv('zoo1.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
y=data['1.7']
print(y)
```

```python
y=np.array(y)
print(y)
```

```python
x=data.drop(columns=['1.7'])
print(x)
```

```python
x=np.array(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
inst=DecisionTreeClassifier()
inst=inst.fit(X_train,Y_train)
Y_pred=inst.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn import metrics
from sklearn.metrics import precision_recall_fscore_support
```

```python
acc=metrics.accuracy_score(Y_test,Y_pred)
acc
```

```python
print("precision,recall,fscore",precision_recall_fscore_support(Y_test,Y_pred,average='macro'))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(inst,fontsize=10)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,Y_pred))
```
